```python
import numpy as np 
import matplotlib.pyplot as plt 
import astropy.constants as const
import astropy.units as u
```

```python
const.c
```

```python
u.amp
```

```python
def theta(wavelength,baseline): # in meters
    return 2.2 * u.rad * wavelength / baseline  # in radians
def time_delay(wavelength,baseline): # in meters
    return baseline * u.m * np.cos(theta(wavelength,baseline)) / const.c # in seconds
def omega(wavelength): # in meters
    return 2 * np.pi * const.c / (wavelength * u.m) #in inverse seconds
def response(wavelength,baseline):
    w = (omega(wavelength)).value
    t = time_delay(wavelength,baseline).value

    return np.cos(w * t)
    
```

```python
baseline = np.arange(0,1000,44)
#baseline = np.arange(0,1000,1)
Response = response(2e-6,baseline)

plt.figure(figsize=(10,10))
plt.plot(baseline,Response)
```
