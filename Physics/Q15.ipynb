```python
import matplotlib.pyplot as plt 
import numpy as np 
import seaborn as sns
from matplotlib import rcParams
sns.set()
rcParams['font.family'] = 'serif'
rcParams['font.sans-serif'] = ['Times']
```

```python
def corr(theta):
    
    x = (1-np.cos(theta)) / 2 
    
    return x * np.log(x) - x / 6 + 1/3
```

```python
angle = np.arange(1,180) 

rad = angle * np.pi / 180 

correlation = corr(rad)

plt.figure(figsize=(10,8))

plt.plot(angle,correlation,lw=10,c="navy",alpha=.8)
plt.ylabel("Arrival Time Correlation",fontsize=20)
plt.xlabel("Angle between Pulsars",fontsize=20)
plt.title("Hellings & Down Curve",fontsize=30)
plt.savefig("HD_Curve.png",transparent=True)
```
