```python
def R(M):
    return M ** (-1/3)
```

```python
import matplotlib.pyplot as plt 
import numpy as np 
import seaborn as sns; sns.set()
```

```python
Mrange = np.arange(0.01,1.4,0.01)
plt.figure(figsize=(8,5))
plt.scatter(Mrange,R(Mrange),c="blue")
plt.xlabel("Mass")
plt.ylabel("Radius")
plt.axvline(1.4,c="blue")
plt.axvline(0.07,c="brown")
plt.text(1.35,2.1,"WD Mass Limit",c='blue',rotation=90)
plt.text(-0.01,1,"Brown Dwarf: 0.01 - 0.07 Msun",rotation=90,c="brown")
```
