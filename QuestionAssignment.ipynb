```python
import pandas as pd
import random
import numpy as np 
```

```python
df = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vRCneqeupUDDN37wG0tErpqjqHxbY94mj6gF4X9d3b1bc4dBCj55AIuUid5JXIyvHL3jNthZm-V_N_i/pub?gid=0&single=true&output=csv")
df = df.replace(np.nan, '', regex=True)
df
```

```python
# Personalized Details
aarya_df = df[(df['Type']=='Base') | (df['Type']=='Radiation')]
bethany_df = df.loc[[0,1,3,4,5,6,7,8,9,10,11,17,18,19,20,21,22,23,24]]
harrison_df = df[(df['Type']=='Base')]
martine_df = df[(df['Type']=='Base')]
nathaniel_df = df[(df['Type']=='Base') | (df['Type']=='Radiation')]

def GenerateQ (df,personalized_df,name):
    # Exclude if already completed
    q_df = personalized_df[~personalized_df["CompletedBy"].str.contains(name)]


    # Pick a question at random
    q_index = random.randrange(q_df.shape[0])
    assigned_no = q_df['Question_No'].iloc[q_index]

    print(assigned_no)
    # Assign to individual
    df['AssignedNo1'][df['Question_No'] == assigned_no] = name

    # Exclude Question That Was Just Assigned 
    q2_df = q_df[~q_df['AssignedNo1'].str.contains(name)]
    q2_index = random.randrange(q2_df.shape[0])

    # Assign Second Question
    assigned_no_2 = q2_df['Question_No'].iloc[q2_index]
    df['AssignedNo2'][df['Question_No'] == assigned_no_2] = name

    # Exclude Question That Was Just Assigned 
    q3_df = q2_df[~q2_df['AssignedNo2'].str.contains(name)]
    q3_index = random.randrange(q3_df.shape[0])

    # Assign Second Question
    assigned_no_3 = q3_df['Question_No'].iloc[q3_index]
    df['AssignedNo3'][df['Question_No'] == assigned_no_3] = name

GenerateQ(df,aarya_df,"Aarya")
GenerateQ(df,bethany_df,"Bethany")
GenerateQ(df,harrison_df,"Harrison")
GenerateQ(df,martine_df,"Martine")
GenerateQ(df,nathaniel_df,"Nathaniel")
df
```
