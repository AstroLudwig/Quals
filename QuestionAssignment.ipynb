```python
import pandas as pd
import random
import numpy as np 
```

```python
df = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vRCneqeupUDDN37wG0tErpqjqHxbY94mj6gF4X9d3b1bc4dBCj55AIuUid5JXIyvHL3jNthZm-V_N_i/pub?gid=0&single=true&output=csv")
df = df.replace(np.nan, '', regex=True)
df
```

```python
# Personalized Details
aarya_df = df[(df['Type']=='Base') | (df['Type']=='Radiation')]
bethany_df = df.loc[[0,1,3,4,5,6,7,8,9,10,11,17,18,19,20,21,22,23,24]]
harrison_df = df[(df['Type']=='Base')]
martine_df = df[(df['Type']=='Base')]
nathaniel_df = df[(df['Type']=='Base') | (df['Type']=='Radiation')]

def GenerateQ (df,personalized_df,name):
  
    # Remove questions that have already been completed
    notcompleted_df = personalized_df[~personalized_df["CompletedBy"].str.contains(name)]
    
    col =  ['AssignedNo1','AssignedNo2','AssignedNo3'] 
    
    # Choose questions not currently being done by someone else    
    notcompleted_notassigned_df = notcompleted_df[(df[col[0]].str.len() == 0) & (df[col[1]].str.len() == 0) & (df[col[2]].str.len() == 0)]
    
    # Pick 3 random values
    assigned_index = random.sample(range(0, len(notcompleted_notassigned_df)), 3)

    # Pick from available questions using random values
    assigned_question = notcompleted_notassigned_df['Question_No'].iloc[assigned_index]
    
    # Assign Questions 
    for question,column in zip(assigned_question,col):
               
        df[column][df['Question_No'] == question] = name

                
                
GenerateQ(df,aarya_df,"Aarya")
GenerateQ(df,bethany_df,"Bethany")
GenerateQ(df,harrison_df,"Harrison")
GenerateQ(df,martine_df,"Martine")
GenerateQ(df,nathaniel_df,"Nathaniel")
df        
```
